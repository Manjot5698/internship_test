```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins") # used penguin dataset of seaborn
```

```python
df.head()
```

```python
df.dropna(inplace = True) # removing the null values
```

```python
df.isnull().sum()
```

## count of penguins of each species using count plot

```python
sns.countplot(x = df["species"])
plt.title("Count of Penguins of each species")
plt.show()
```

## Scatterplot to find Relationship between Flipper Length and Body Mass

```python
sns.scatterplot(x ="flipper_length_mm",
               y ="body_mass_g",
               hue = "species",
               data = df)
plt.title("Relationship between Flipper Length and Body Mass")
plt.show()
```

## Distribution of Body Mass

```python
sns.histplot(x = df["body_mass_g"],
            bins = 20,
             kde = True)
plt.title("Distribution of Body Mass")
plt.show()
```

## boxplot for bill length of each species

```python
sns.boxplot(y=df["bill_length_mm"],x ="species",data = df )
plt.title("Bill Length of each species")
plt.show()
```

## Heatmap

```python
numeric_df = df.select_dtypes(include ="number")
sns.heatmap(
    data = numeric_df.corr(),
    annot = True,
    cmap = "coolwarm"
    
)
```
