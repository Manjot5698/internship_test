```python
import pandas as pd
```

```python
df = pd.read_csv("employee_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.drop(columns =["Name"],axis = 1,inplace = True)
```

## Handling Missing values

```python
df["Department"] = df["Department"].fillna("unknown")
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"])
```

```python
df["JoiningDate"] = df["JoiningDate"].fillna(df["JoiningDate"].median())
```

```python
df["Status"] = df["Status"].fillna("unknown")
```

## Employees that are no longer working

```python
df["Status"].unique()
```

```python
resigned = df[df["Status"] == 'Resigned']
print(resigned)
```

## Department wise average salary of employees

```python
avg_salary = df.groupby("Department")["Salary"].mean()
print(avg_salary)
```

## Employees joined after 2020

```python
recent = df[df["JoiningDate"].dt.year>2020]
print(recent)
```

```python
df.to_csv("cleaned_dataframe.csv",index =False)
```
